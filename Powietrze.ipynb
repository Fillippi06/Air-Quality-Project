```python
def transform_data(df):
    df = df.iloc[4:]
    
    df.columns = df.iloc[0]
    df = df[1:]

    df = df.filter(like='PmGd')

    return df
```

```python
df_co = transform_data(df_co)
df_no2 = transform_data(df_no2)
df_o3 = transform_data(df_o3)
df_pm10 = transform_data(df_pm10)
df_pm25 = transform_data(df_pm25)
df_so2 = transform_data(df_so2)
```
