```python
def transform_data(df):
    """Transformuje dane o jakości powietrza, usuwając pierwsze 4 wiersze i filtrując kolumny.
    
    Args:
        df (pd.DataFrame): Wejściowy DataFrame z danymi nt. jakości powietrza.
    
    Returns:
        pd.DataFrame: Przetworzony DataFrame.
    """
    # Usunięcie pierwszych 4 wierszy
    df = df.iloc[4:]

    # Ustawia pierwszy wiersz jako nagłówek
    df.columns = df.iloc[0]
    df = df[1:]

    # Wybiera kolumny, które zawierają 'PmGd' w nazwie
    df = df.filter(like='PmGd')

    # Zwraca DataFrame
    return df
```
